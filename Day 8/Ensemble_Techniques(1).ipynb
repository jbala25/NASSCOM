#Ensemble Techniques

Ensemble techniques are methods that combine multiple models (often called "weak learners") to produce a stronger, more accurate model.

#✅ Why Use Ensemble Methods?

> Improve accuracy and robustness

> Reduce overfitting

> Handle bias-variance trade-off

> Perform better than individual models in most real-world tasks

#🔹 Types of Ensemble Techniques:

1. Bagging (Bootstrap Aggregating)
Trains multiple models independently on random subsets of the data (with replacement)

Final prediction is usually made by voting (classification) or averaging (regression)

✅ Reduces variance, good for high-variance models like decision trees

🛠️ Example:




```python
from sklearn.ensemble import BaggingClassifier

#Popular Method:
#⭐ Random Forest = Bagging + Decision Trees + Feature Randomness
```

#2. Boosting

> Trains models sequentially

> Each model tries to correct the errors of the previous one

> Final prediction is a weighted sum of all models

✅ Reduces bias, great for weak learners

🛠️ Examples:

```python
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
```

#Popular Boosting Algorithms:

🔸 AdaBoost

🔸 Gradient Boosting

🔸 XGBoost

🔸 LightGBM

🔸 CatBoost

#3. Stacking (Stacked Generalization)

Trains multiple different models (base learners)

>Combines their outputs using a meta-model (often logistic regression)

✅ Leverages diverse algorithms to improve performance

🛠️ Example:

```python
from sklearn.ensemble import StackingClassifier
```

#Summary Table:

| Technique    | Approach            | Goal              | Famous Model       |
| ------------ | ------------------- | ----------------- | ------------------ |
| **Bagging**  | Parallel Training   | Reduce Variance   | Random Forest      |
| **Boosting** | Sequential Training | Reduce Bias       | XGBoost, AdaBoost  |
| **Stacking** | Layered Models      | Combine Strengths | StackingClassifier |


#Practical hands -on

#Bagging
Fifty decision trees are trained in parallel on bootstrap samples, then their votes are averaged.

#Boosting (AdaBoost)
Fifty shallow trees are trained sequentially, each one paying extra attention to the points the previous trees mis‑classified.

#Stacking
A decision tree and an SVM each make predictions; a logistic‑regression “meta‑model” learns to blend their outputs.

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# 1. Load a small sample dataset (Iris)
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

# 2. Bagging (Bootstrap Aggregating) with Decision Trees
bagging = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=50,
    random_state=42
)
bagging.fit(X_train, y_train)
bagging_acc = accuracy_score(y_test, bagging.predict(X_test))

```

```python
# 3. Boosting (AdaBoost) with 50 weak learners
boosting = AdaBoostClassifier(n_estimators=50, random_state=42)
boosting.fit(X_train, y_train)
boosting_acc = accuracy_score(y_test, boosting.predict(X_test))

```

```python
# 4. Stacking with two base models & a Logistic Regression meta‑model
stacking = StackingClassifier(
    estimators=[
        ("dt", DecisionTreeClassifier()),
        ("svc", SVC(probability=True))
    ],
    final_estimator=LogisticRegression(max_iter=1000),
    cv=5
)
stacking.fit(X_train, y_train)
stacking_acc = accuracy_score(y_test, stacking.predict(X_test))

```

```python
# 5. Summarize the results
results_df = pd.DataFrame(
    {
        "Technique": ["Bagging", "AdaBoost (Boosting)", "Stacking"],
        "Accuracy": [bagging_acc, boosting_acc, stacking_acc]
    }
)

# 6. Display the results as an interactive table
display(results_df)
```
